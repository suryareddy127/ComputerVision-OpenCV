## Annotating images
-  image annotation means adding text, shapes, or markers on top of an image. This is super useful in computer vision (e.g., labeling bounding boxes, highlighting features) or just for presentation.


```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('own-sketch-to-the-image.webp')

# Draw a rectangle (x1,y1) to (x2,y2)
cv2.rectangle(img, (50, 50), (200, 200), (0, 255, 0), 2)

# Add text
cv2.putText(img, 'Object', (60, 45), cv2.FONT_HERSHEY_SIMPLEX, 
            0.8, (255, 0, 0), 2, cv2.LINE_AA)

# Convert BGR  RGB for Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

###  Common Arithmetic Operations on Images
1. Addition (Brightness Enhancement)
- Adding a constant value to all pixels increases brightness.
- Increase brightness by adding 50

2. Subtraction (Darkening or Background Removal)
- Subtracting reduces brightness or removes background intensity.

3. Multiplication (Contrast Enhancement)
- Multiplying pixel values scales intensity, enhancing contrast.

4. Division (Normalization)
- Dividing reduces intensity, useful for normalization.


 Key Notes
- Arithmetic operators are applied pixel by pixel.
- Always use OpenCV functions (cv2.add, cv2.subtract, etc.) instead of raw NumPy operators to avoid overflow/underflow issues.
- These methods are global enhancements (apply to the whole image).



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('your_image.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
bright_img = cv2.add(img, 50)

plt.imshow(bright_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
dark_img = cv2.subtract(img, 50)

plt.imshow(dark_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
contrast_img = cv2.multiply(img, 1.5)  # 1.5x contrast

plt.imshow(contrast_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
normalized_img = cv2.divide(img, 2)

plt.imshow(normalized_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
enhanced_img = cv2.add(cv2.multiply(img, 1.2), 30)  # increase contrast + brightness
plt.imshow(enhanced_img, cmap='gray')
plt.axis('off')
plt.show()
```

## image filtering


 Types of Image Filtering
1. Smoothing / Blurring (Noise Reduction)
- Reduces noise and detail.
2. Sharpening (Edge Enhancement)
- Highlights edges and fine details
3. Edge Detection
- Detects boundaries and shapes.


```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('own-sketch-to-the-image.webp')

# Average blur
blur = cv2.blur(img, (5,5))

# Gaussian blur
gaussian = cv2.GaussianBlur(img, (5,5), 0)

# Median blur
median = cv2.medianBlur(img, 5)

plt.subplot(1,3,1); plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)); plt.title("Average")
plt.subplot(1,3,2); plt.imshow(cv2.cvtColor(gaussian, cv2.COLOR_BGR2RGB)); plt.title("Gaussian")
plt.subplot(1,3,3); plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB)); plt.title("Median")
plt.show()
```

```python
import numpy as np

# Sharpening kernel
kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

sharpened = cv2.filter2D(img, -1, kernel)

plt.imshow(cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB))
plt.title("Sharpened")
plt.axis('off')
plt.show()
```

```python
# Sobel filters
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)

# Canny edge detector
edges = cv2.Canny(img, 100, 200)

plt.subplot(1,3,1); plt.imshow(sobelx, cmap='gray'); plt.title("Sobel X")
plt.subplot(1,3,2); plt.imshow(sobely, cmap='gray'); plt.title("Sobel Y")
plt.subplot(1,3,3); plt.imshow(edges, cmap='gray'); plt.title("Canny")
plt.show()
```

### Common Morphological Operations in OpenCV

1. Erosion
- Removes pixels at object boundaries.
- Useful for removing small white noise.

2. Dilation
- Adds pixels to object boundaries.
- Useful for filling small holes.

3. Opening (Erosion  Dilation)
- Removes noise while preserving shape.

4. Closing (Dilation  Erosion)
- Fills small holes inside objects.

5. Morphological Gradient
- Difference between dilation and erosion.
- Highlights object boundaries.

 Key Notes
- Kernel (structuring element) defines the neighborhood (e.g., 33, 55).
- Operations are applied pixel by pixel relative to the kernel.
- Best used on binary images (thresholded).


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('binary_image.png', 0)  # load grayscale/binary image

kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(img, kernel, iterations=1)

plt.imshow(erosion, cmap='gray')
plt.title("Erosion")
plt.axis('off')
plt.show()
```

```python
dilation = cv2.dilate(img, kernel, iterations=1)

plt.imshow(dilation, cmap='gray')
plt.title("Dilation")
plt.axis('off')
plt.show()
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.imshow(opening, cmap='gray')
plt.title("Opening")
plt.axis('off')
plt.show()
```

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.imshow(closing, cmap='gray')
plt.title("Closing")
plt.axis('off')
plt.show()
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

plt.imshow(gradient, cmap='gray')
plt.title("Gradient")
plt.axis('off')
plt.show()
```
